### Getting data from twitter using #appleevent and loading into pickle files:

```python
import tweepy
import csv
import pickle

from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_secret)

api = tweepy.API(auth)


apple_tweets = list(tweepy.Cursor(api.search,q="#AppleEvent",lang="en").items(1500))

with open('apple_tweets.pkl', 'wb') as f:
    pickle.dump(apple_tweets,f)
    
with open('apple_tweets.pkl', 'rb') as f:
    apple_tweets=pickle.load(f)
```

## Performing the Sentiment Analysis

```python
from nltk_helpers import get_sentiments
    
#print(get_sentiments(michelle_tweets[0].text))

for tweet in apple_tweets:
    print("Apple tweets : ",get_sentiments(tweet.text))
print()

sentiments = [get_sentiments(tweet.text) for tweet in apple_tweets]
```

### Finding out the postive, negative and neutral tweets by considering the compound score values.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

positive = []
negative = []
neutral = []

# Calculating if the tweet is postive, negative or neutral by using compound scores.
# If the compound score is greater than 0.5 the tweet is considered as positive;
# Else If the compund Score is less than -0.5 the tweet is considered as negative;
# Otherwise, the tweet is neural.
for tweet in apple_tweets:   
    if(analyzer.polarity_scores(tweet.text)['compound']) >=0.5:        
        positive.append(tweet.text)
    elif(analyzer.polarity_scores(tweet.text)['compound']) <= -0.5:        
        negative.append(tweet.text)
    else:
        neutral.append(tweet.text)        

print()

## Calculating the percentage of respective sentiment among the total tweets.
total=len(positive)+len(negative)+len(neutral)
per_pos=(len(positive)/total)*100
per_neg=(len(negative)/total)*100
per_neu=(len(neutral)/total)*100
print('length of positive tweets: ',per_pos)
print('length of negative tweets: ',per_neg)
print('length of neutral tweets: ',per_neu)
```

## Visualizing the Sentiment Analysis using Bar Graph

```python
# Considering the Polarities(postive, negative, neutral) on the X-axis and Percentages of the tweets on Y axis to plot a bar graph.

import matplotlib.pyplot as plt
heights = [per_pos,per_neg,per_neu]
print(heights)


left=[1,2,3]

tick_label=['Positive','Negative', 'Neutral']


plt.bar(left,heights,tick_label=tick_label, width=0.2, color=['maroon'])

plt.xlabel('Polarities')

plt.ylabel('Percentage')
plt.title('Reactions on the apple event based on Compound Score')


plt.show()
```

From the above bar graph we can observe that there are more neutral responses than positive and negative, among the tweets

## Lexical Diversity:

```python
def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)

apple_diversities = []
for tweet in apple_tweets:
    apple_diversities.append(lexical_diversity(tweet.text))
#print("Diversities for apple - ",apple_diversities)

print("Minimum Diverisities - ", min(apple_diversities))
print("Maximum Diverisities - ", max(apple_diversities))
```

From the above values, we can understand that many synonyms are used to express their opinions or responses in the tweets

## Calculating mentions of the Apple products among the tweets:


```python
iPhoneX=0
AppleWatch=0
iPhone=0



for tweet in apple_tweets:        
    if "#iPhoneX" in tweet.text or "iPhone X" in tweet.text or "iPhoneX" in tweet.text: 
        iPhoneX+=1
    elif "#applewatch" in tweet.text.lower() or "apple watch" in tweet.text.lower() or "applewatch" in tweet.text.lower():
        AppleWatch+=1
    elif "#iPhone" in tweet.text.lower() or "iphone" in tweet.text.lower() or "iPhone" in tweet.text.lower():
        iPhone+=1     

print('Tweets on iphone x : ',iPhoneX)
print('Tweets on Apple watch : ',AppleWatch)
print('Tweets on iphones:', iPhone)
```

## Visualizing which Apple Product has the most attention:

```python
import matplotlib.pyplot as plt
heights = [iPhoneX,AppleWatch,iPhone]
print(heights)


left=[1,2,3]

tick_label=['iPhoneX','AppleWatch','iPhone']


plt.bar(left,heights,tick_label=tick_label, width=0.2, color=['grey'])

plt.xlabel('Apple Products')

plt.ylabel('Number tweets on Apple Products')
plt.title('The Apple Product that has the most attention')


plt.show()
```

From the above graph, we see that there are more number of tweets mentioning about the apple watch compared to the iPhone X or other iPhones.

## Finding the Commonly used words:

```python
from collections import Counter
words = []
for tweet in apple_tweets: 
    words += tweet.text.lower().split()

c=Counter(words)

print("Most common words for Apple event: \n",c.most_common(40))
print()
```

## Visualizing the most frequently occurred words among the tweets:


```python
import matplotlib.pyplot as plt

NUM_VALUES = 20
heights = list(c.values())

heights.sort(reverse=True)


plt.bar(range(NUM_VALUES),heights[:NUM_VALUES],color=['goldenrod'])



plt.title('Word Occurrences')
plt.ylabel('# occurrences')
plt.xticks(rotation=90)
keys =list(c.keys())
keys.sort(reverse=True,key=lambda x:c[x])
#print(keys)
plt.xticks(range(NUM_VALUES),keys)
plt.show()



```

From the above bar graph we can see that the most commonly used words in the tweets are #applewatch, apple, #wwdc2018 and so on.
